```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
np.random.seed(0)
time_series = np.random.randint(0, 100, size=50)
```

```python
alpha = 0.2
```

```python
model = ExponentialSmoothing(time_series, trend=None, seasonal=None, initialization_method="heuristic")
smoothed_model = model.fit(smoothing_level=alpha, optimized=False)
smoothed_series = smoothed_model.fittedvalues
```

```python
plt.plot(time_series, label='Original series')
plt.plot(smoothed_series, label='Smoothed series')
plt.legend()
plt.title('Exponential Smoothing with Holt-Winters Method')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
forecast_steps = 10
forecast = smoothed_model.forecast(steps=forecast_steps)
```

```python
plt.plot(time_series, label='Original series')
plt.plot(smoothed_series, label='Smoothed series')
plt.plot(range(len(time_series), len(time_series) + forecast_steps), forecast, label='Forecasted values', linestyle='--')
plt.legend()
plt.title('Exponential Smoothing with Holt-Winters Method')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
# double exponential smoothing
alpha = 0.2  
beta = 0.1  
```

```python
model = ExponentialSmoothing(time_series, trend='add', seasonal=None, initialization_method="heuristic")
smoothed_model = model.fit(smoothing_level=alpha, smoothing_slope=beta, optimized=False)
smoothed_series = smoothed_model.fittedvalues
```

```python
forecast_steps = 10
forecast = smoothed_model.forecast(steps=forecast_steps)
```

```python
plt.plot(time_series, label='Original series')
plt.plot(smoothed_series, label='Smoothed series')
plt.plot(range(len(time_series), len(time_series) + forecast_steps), forecast, label='Forecasted values', linestyle='--')
plt.legend()
plt.title('Double Exponential Smoothing (Holt\'s Method)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
# triple exponential smoothing 
alpha = 0.2 
beta = 0.1   
gamma = 0.1  
```

```python
model = ExponentialSmoothing(time_series, trend='add', seasonal='add', seasonal_periods=4, initialization_method="heuristic")
smoothed_model = model.fit(smoothing_level=alpha, smoothing_slope=beta, smoothing_seasonal=gamma, optimized=False)
smoothed_series = smoothed_model.fittedvalues
```

```python
forecast_steps = 10
forecast = smoothed_model.forecast(steps=forecast_steps)
```

```python
plt.plot(time_series, label='Original series')
plt.plot(smoothed_series, label='Smoothed series')
plt.plot(range(len(time_series), len(time_series) + forecast_steps), forecast, label='Forecasted values', linestyle='--')
plt.legend()
plt.title('Triple Exponential Smoothing (Holt-Winters\' Method)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
